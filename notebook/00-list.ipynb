```python
from sympy import *
init_printing()
```

```python
# Assignment 1
v1 = 5 # m/s
v2 = 2 # m/s

x = symbols('x')
y = 50 - x

s1 = sqrt(20 ** 2 + y ** 2)
s2 = sqrt(20 ** 2 + x ** 2)

t1 = s1 / v1
t2 = s2 / v2
```

```python
t = t1 + t2

plot(diff(t), (x, 0, 50))
plot(t, (x, 0, 50))
```

```python
x0 = solve(diff(t))[0]
x0.evalf()
```

```python
t.subs(x, x0).evalf()
```

```python
# Assignment 2

m = 1.7 # wysokosc oczu
f = 0.7 # wysokosc sukienki

# x     - odleglosc miedzy panem a pania
# alpha - najwiekszy kat (szukana)
x = symbols('x') 

beta = atan((m - f) / x)
gamma = atan((x / m))

alpha = 90 - beta - gamma
plot(alpha, (x, 0, 100))

```

```python
alpha
```

```python
alphap = diff(alpha)
plot(alphap, (x, 0, 100))

guess = solve(alphap)
guess
```

```python
alphap
```

```python
solve(alphap > 0)
```

```python
a = alpha.subs(x, guess[1])
a
```
