```python
from sympy import *
init_printing()
```

```python
x = symbols('x')
e = exp(1)
```

```python
# Funkcja
f = e ** atan(x)

plot(f)

A = limit(f / x, x, oo)
B = limit(f - A * x, x, oo)
Bp = limit(f - A * x, x, -oo)

y = A * x + B

print(f'A = {A}, B = {B}, B\' = {Bp}')
print(f.subs(x, 0), f.subs(x, 1), round(f.subs(x, 2), 2), round(f.subs(x, -1), 5))
```

```python
# 3

xs = [1.11, 2.22, 3.33, 4.44]

k = 1
while k < 5:
    prim = "'" * k
    p = round(diff(f, x, k).subs(x, xs[k - 1]), 3)
    print(f'f{prim}(x) = {p}')
    k += 1
```

```python
# 4
fp = diff(f, x)
plot(fp)
solve(fp > 0), solve(fp < 0)
```

```python
fpp = diff(f, x, 2)
plot(fpp)
solve(fpp > 0), solve(fpp < 0)
```
